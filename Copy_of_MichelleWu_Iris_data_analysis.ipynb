```python
import pandas as pd

df = pd.read_csv('iris.csv')
print(df.head())  # View the first few rows
print(df.info())  # Get info about data types, missing values, etc.
```

```python
print(df.describe())  # Summary statistics (mean, std, quartiles, etc.)
```

```python
print(df.columns)
sepal_length = df['sepal.length']
sepal_width = df['sepal.width']
```

```python
large_sepal_length = df[sepal_length > 5]
print(large_sepal_length)
```

```python
# Describe numerical columns (Sepal Length, Sepal Width, Petal Length, Petal Width)
import matplotlib.pyplot as plt
print(df.describe())

# Check for missing values
print(df.isnull().sum())  # Shows how many missing values per column

# Distribution of Sepal Length across Iris species
plt.figure(figsize=(8, 6))
df.groupby('variety')['sepal.length'].hist(alpha=0.7, bins=20, label='Sepal Length')
plt.legend()
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Number of Iris')
plt.title('Sepal Length Distribution by Species')
plt.show()
```

```python
# Distribution of Sepal Length across Iris species
plt.figure(figsize=(8, 6))
df.groupby('variety')['sepal.width'].hist(alpha=0.7, bins=20, label='Sepal Width')
plt.legend()
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Number of Iris')
plt.title('Sepal Width Distribution by Species')
plt.show()
```

```python
# Distribution of Petal Length across Iris species
plt.figure(figsize=(8, 6))
df.groupby('variety')['petal.length'].hist(alpha=0.7, bins=20, label='Patel Length')
plt.legend()
plt.xlabel('Petal Length (cm)')
plt.ylabel('Number of Iris')
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
# Distribution of Petal Width across Iris species
plt.figure(figsize=(8, 6))
df.groupby('variety')['petal.width'].hist(alpha=0.7, bins=20, label='Patel Width')
plt.legend()
plt.xlabel('Petal Width (cm)')
plt.ylabel('Number of Iris')
plt.title('Petal Width Distribution by Species')
plt.show()
```
